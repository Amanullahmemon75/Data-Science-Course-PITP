# Matplotlib  Introduction

**Step 1: Define data**

```python
# X-axis values
x = [1, 2, 3, 4, 5]
# Y-axis values
y = [2, 4, 6, 8, 10]
```

```python
plt.plot(x, y)  # Basic line plot
plt.title("Basic Line Plot")  # Add a title
plt.xlabel("X-axis")  # Label for the X-axis
plt.ylabel("Y-axis")  # Label for the Y-axis
plt.show()  # Show the plot
```

**Step 3: Adding a label, color, and linestyle**

```python
plt.plot(x, y, label="Linear Growth", color="black", linestyle='--' )
plt.title("Line Plot with Label and Color")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # Add a legend
plt.show()  # Show the plot
```

**Step 4: Adding markers**

```python
plt.plot(x, y, label="Linear Growth", color="blue", linestyle="-", marker="o", markersize=8, markerfacecolor="yellow", markeredgecolor="black")
plt.title("Line Plot with Markers")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()  # Show the plot
```

**Step 5: Adding gridlines**

```python
plt.plot(x, y, label="Linear Growth", color="blue", linestyle="-", marker="o", markersize=8, markerfacecolor="yellow", markeredgecolor="black")
plt.title("Line Plot with Gridlines")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)  # Show gridlines
plt.legend()
plt.show()  # Show the plot
```

**Step 6: Setting limits for axes**

```python
plt.plot(x, y, label="Linear Growth", color="blue", linestyle="-", marker="o", markersize=8, markerfacecolor="yellow", markeredgecolor="black")
plt.title("Line Plot with Axis Limits")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.xlim(0, 6)  # Set the limits for the X-axis
plt.ylim(0, 12)  # Set the limits for the Y-axis
plt.grid(True)
plt.legend()
plt.show()  # Show the plot
```

**Step 7: Customizing tick marks**

```python
plt.plot(x, y, label="Linear Growth", color="blue", linestyle="-", marker="o", markersize=8, markerfacecolor="yellow", markeredgecolor="black")
plt.title("Line Plot with Custom Tick Marks")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.xlim(0, 6)
plt.ylim(0, 12)
plt.xticks(ticks=[1, 2, 3, 4, 5], labels=['One', 'Two', 'Three', 'Four', 'Five'])  # Custom tick labels for X-axis
plt.yticks(ticks=[2, 4, 6, 8, 10], labels=['Low', 'Med-Low', 'Medium', 'Med-High', 'High'])  # Custom tick labels for Y-axis
plt.grid(True)
plt.legend()
plt.show()  # Show the plot
```

```python
import matplotlib.pyplot as plt

# Step 1: Define data
# X-axis values
x = [1, 2, 3, 4, 5]
# Y-axis values
y = [2, 4, 6, 8, 10]

# Step 2: Create a basic line plot with customizations
plt.plot(x, y,
         label="Linear Growth",        # Add a label for the line (this will appear in the legend)
         color="blue",                 # Set the color of the line (can be any color like 'red', 'green', etc.)
         linestyle="-.",                # Set the style of the line ('-', '--', '-.', ':')
         marker="o",                   # Set the marker for each data point ('o', 's', '^', etc.)
         markersize=8,                 # Size of the marker
         markerfacecolor="yellow",      # Color inside the marker
         markeredgecolor="black")       # Color around the marker

# Step 3: Adding a title and axis labels
plt.title("Customized Line Plot")       # Add title to the plot
plt.xlabel("X-axis (Time)")             # Label for the X-axis
plt.ylabel("Y-axis (Value)")            # Label for the Y-axis

# Step 4: Add gridlines for better readability
plt.grid(True)                          # Show gridlines on the plot
plt.grid(color='gray', linestyle='--', linewidth=0.5)  # Customizing gridlines

# Step 5: Adding a legend to identify the line
plt.legend(loc="upper left")            # Place the legend in the upper left corner

# Step 6: Customizing the X and Y limits (optional)
plt.xlim(0, 6)                          # Set the limits for the X-axis
plt.ylim(0, 12)                         # Set the limits for the Y-axis

# Step 7: Customizing tick marks (optional)
plt.xticks(ticks=[1, 2, 3, 4, 5], labels=['One', 'Two', 'Three', 'Four', 'Five'])  # Custom tick labels for X-axis
plt.yticks(ticks=[2, 4, 6, 8, 10], labels=['Low', 'Med-Low', 'Medium', 'Med-High', 'High'])  # Custom tick labels for Y-axis

# Step 8: Display the plot
plt.show()                              # Show the plot on the screen
```

# Equations

**1st method by creating function**

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function y(x) = x^2 + 2x - 10
def y(x):
    return x**2 + 2*x - 10

# Generate x values using np.linspace to create 200 evenly spaced points between -10 and 10
x = np.linspace(-10, 10, 200)

# Calculate the corresponding y values by passing x to the function
y_values = y(x)

# Create a plot figure with specified size
plt.figure(figsize=(4, 3))

# Plot x vs y_values, label the curve, and choose a color
plt.plot(x, y_values, label='y(x) = x^2 + 2x - 10', color='blue')

# Add title and labels for the x and y axes
plt.title('Graph of Quadratic Function')
plt.xlabel('x')  # Label for the x-axis
plt.ylabel('y')  # Label for the y-axis

# Add grid lines to the plot for better visibility
plt.grid(True)

# Show the legend that contains the label defined earlier
plt.legend()

# Display the plot
plt.show()
```

**2nd method using the formula/equation directly**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values using np.arange to create integer points between -10 and 10 (inclusive)
x = np.arange(-10, 11, 1)
# Or
# Generate x values using np.linspace to create 400 evenly spaced points between -10 and 10
# x = np.linspace(-10, 10, 200)

# Calculate y values using the formula directly
y = x**2 + 2*x - 10

# Create a plot figure with a specified size
plt.figure(figsize=(8, 6))

# Plot x vs y, with markers at each point ('o'), a solid line ('-'), and green color
plt.plot(x, y, marker='o', linestyle='-', color='green', label='y = x^2 + 2x - 10')

# Add title and labels for the x and y axes
plt.title('Graph of Quadratic Equation')
plt.xlabel('x')  # Label for the x-axis
plt.ylabel('y')  # Label for the y-axis

# Add grid lines to the plot for clarity
plt.grid(True)

# Show the legend that contains the label defined earlier
plt.legend()

# Display the plot
plt.show()
```

**Linear Eqn**

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function h(t) = 11t - 10
def h(t):
    return 11*t - 10

# Generate t values from -10 to 10
t = np.linspace(-10, 10, 400)

# Calculate the corresponding h(t) values
h_values = h(t)

# Create a plot figure with a specified size
plt.figure(figsize=(8, 6))

# Plot t vs h(t), and label the curve
plt.plot(t, h_values, label='h(t) = 11t - 10', color='red')

# Add title and labels for the t and h axes
plt.title('Graph of Linear Equation')
plt.xlabel('t')  # Label for the t-axis
plt.ylabel('h(t)')  # Label for the h(t)-axis

# Add grid lines to the plot for better visibility
plt.grid(True)

# Show the legend that contains the label defined earlier
plt.legend()

# Display the plot
plt.show()
```

**Cubic Eqn**

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function p(q) = 0.5q^3 + 3q^2 - 5q + 5
def p(q):
    return 0.5*q**3 + 3*q**2 - 5*q + 5

# Generate q values from -10 to 10
q = np.linspace(-10, 10, 400)

# Calculate the corresponding p(q) values
p_values = p(q)

# Create a plot figure with a specified size
plt.figure(figsize=(8, 6))

# Plot q vs p(q), and label the curve
plt.plot(q, p_values, label='p(q) = 0.5q^3 + 3q^2 - 5q + 5', color='purple')

# Add title and labels for the q and p(q) axes
plt.title('Graph of Cubic Equation')
plt.xlabel('q')  # Label for the q-axis
plt.ylabel('p(q)')  # Label for the p(q)-axis

# Add grid lines to the plot for better visibility
plt.grid(True)

# Show the legend that contains the label defined earlier
plt.legend()

# Display the plot
plt.show()
```

# Seaborn Introduction

**Introduction to Seaborn:** Seaborn is a data visualization library built on top of Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. While Matplotlib offers extensive customization and control over plot appearance, Seaborn simplifies the process of creating complex visualizations with less code and better default aesthetics(the appearance or visual appeal of something), especially for statistical data.

**Advantages:**
1. Simplicity: Seaborn simplifies many tasks that would require more code in Matplotlib.
Aesthetics: Seaborn provides better-looking plots by default.
2. Statistical Capabilities: Seaborn excels in statistical visualization, making it a great choice for data exploration.
3. Pandas Integration: Works effortlessly with Pandas DataFrames, allowing for easy and intuitive mapping of data to visualizations.
4. Faceting and Aggregation: Seaborn makes it easy to create faceted plots and automatically handles aggregations for statistical summaries.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Line Plot with Seaborn**

```python
# Step 1: Create sample data
data = {
    'Time': [1, 2, 3, 4, 5],
    'Value': [2, 4, 6, 8, 10]
}

# Step 2: Create a DataFrame
df = pd.DataFrame(data)

# Step 3: Create a line plot
sns.lineplot(x='Time', y='Value', data=df, marker='o', color='blue')

# Step 4: Add title and labels
plt.title("Line Plot using Seaborn")          # Title of the plot
plt.xlabel("Time")                             # Label for X-axis
plt.ylabel("Value")                            # Label for Y-axis

# Create a plot figure with specified size
plt.figure(figsize=(2, 2))

# Step 5: Show the plot
plt.show()                                     # Display the line plot
```

**Bar Chart with Seaborn**

```python
# Step 1: Define data
data = {'Products': ['Product A', 'Product B', 'Product C'],
        'Sales': [50, 30, 40]}

# Step 2: Create a DataFrame
df = pd.DataFrame(data)

# Step 3: Create a bar chart
sns.barplot(x='Products', y='Sales', data=df, palette='Blues_d', hue='Products')  # Set color palette

# Step 4: Add title
plt.title("Sales of Products")

# Step 5: Show the plot
plt.show()
```

**Scatter Plot with Seaborn**

```python
# Step 1: Generate random data
data = {'X': np.random.rand(50) * 100, 'Y': np.random.rand(50) * 100}
df = pd.DataFrame(data)

# Step 2: Create a scatter plot
sns.scatterplot(x='X', y='Y', data=df, color='green', marker='o')

# Step 3: Add title
plt.title("Scatter Plot of Random Data")

# Step 4: Show the plot
plt.show()
```

**Histogram with Seaborn**

```python
# Step 1: Generate random data
data = np.random.randn(1000)

# Step 2: Create a histogram
sns.histplot(data, bins=30, color='lightgreen', kde=True)  # kde=True adds a Kernel Density Estimate

# Step 3: Add title
plt.title("Histogram of Random Data")

# Step 4: Show the plot
plt.show()
```

**Plotting Categorical and Continuous Data**

A **box plot** (also known as a box-and-whisker plot) is a type of plot used to display the distribution of a dataset. It shows the summary of a set of data values, including their median, quartiles, and potential outliers. It provides a graphical view of the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum values of a dataset.

**Key Features of a Box Plot:**
1. Box: The box represents the interquartile range (IQR), which is the range between the first quartile (25th percentile) and the third quartile (75th percentile).
2. Line Inside the Box: This represents the median (50th percentile).
3. Whiskers: The lines extending from the box show the minimum and maximum values within 1.5 times the IQR from the quartiles.
4. Outliers: Data points outside the

```python
# Boxplot Example:
# Boxplot
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Boxplot of Total Bill by Day and Gender")
plt.show()
```

A **violin plot** is similar to a box plot but with additional information. It shows the distribution of the data across different categories and also includes a density plot to provide more insights into the data distribution's shape. It combines the features of a box plot and a kernel density plot, giving a clearer view of the distribution.

**Key Features of a Violin Plot:**
1. Density: Each side of the violin shape shows the kernel density estimate of the data, which represents how frequently data points occur.
2. Box Plot Inside: A violin plot often includes a mini box plot inside to represent the median and quartiles.
3. Symmetry: The width of the "violin" indicates the density of the data at different values.

```python
# Violin Plot Example:
# Violin Plot
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.title("Violin Plot of Total Bill by Day and Gender")
plt.show()
```
